```python
import pandas as pd
import numpy as np

file_path = "./epcg21.sas7bdat"

data = []
data = pd.read_sas(file_path) # Reading Initial Data
data = data.applymap(lambda x: x.decode('utf-8') if isinstance(x, bytes) else x) # Cleaning the data (Converts bytes to string)
data.head()
```

```python
#Choosing our Variables

variables = [
    "AGE",  # Age of the person
    "GENDER",  # Gender of the person
    "BTHUS",  # Place of birth (U.S./Non-U.S.)
    "CTZDUAL",  # Dual citizenship indicator
    "CTZUSIN",  # US citizen or not
    "EDDAD",  # Father's education level
    "EDMOM",  # Mother's education level
    "VETSTAT",  # Veteran status
    "RACEM",  # Race
    "BACARN",  # Carnegie classification of institution
    "BSDGN",  # Number of bachelor's or higher degrees
    "DGRDG",  # Highest degree type
    "MRDGRUS",  # Location of the school awarding the most recent degree (U.S./Non-U.S.)
    "MARIND",  # Marital status indicator
    "FACADV",  # Importance of job's opportunities for advancement
    "FACBEN",  # Importance of job's benefits
    "FACCHAL",  # Importance of job's intellectual challenge
    "FACIND",  # Importance of job's degree of independence
    "FACLOC",  # Importance of job's location
    "FACRESP",  # Importance of job's level of responsibility
    "FACSAL",  # Importance of job's salary
    "FACSEC",  # Importance of job's security,
    "LFSTAT" # Y - Variable
]
data = data[variables] # Only keeping our variables
data = data[data['LFSTAT'] != '3'] # Removing people that do not belong to Labor Force
```

```python
data['LFSTAT'] = data['LFSTAT'].replace('2', False) # Changing the encoding on unemployed people to False
data['LFSTAT'] = data['LFSTAT'].replace('1', True) # Changing the encoding on employed people to True
```

```python
data['BACARN'] = data['BACARN'].replace('L', '0') # People with logical skips and missing values imply no bachelor's degree
data['BACARN'] = data['BACARN'].replace('M', '0')
```

```python
data['GENDER'] = data['GENDER'].replace('M', False) # People with logical skips and missing values imply no bachelor's degree
data['GENDER'] = data['GENDER'].replace('F', True)
```

```python
data.head()
```

```python
data.to_csv('Data.csv', index=False) #Saving our Clean Dataset in csv format
```

```python
data['BTHUS'] = data['BTHUS'].replace('Y', True)
data['BTHUS'] = data['BTHUS'].replace('N', False)

data['CTZDUAL'] = data['CTZDUAL'].replace('Y', True)
data['CTZDUAL'] = data['CTZDUAL'].replace('N', False)

data['CTZUSIN'] = data['CTZUSIN'].replace('Y', True)
data['CTZUSIN'] = data['CTZUSIN'].replace('N', False)

data['MRDGRUS'] = data['MRDGRUS'].replace('Y', True)
data['MRDGRUS'] = data['MRDGRUS'].replace('N', False)

data['MARIND'] = data['MARIND'].replace('Y', True)
data['MARIND'] = data['MARIND'].replace('N', False)
```

```python
column = ['EDDAD', 'EDMOM', 'VETSTAT', 'RACEM', 'BACARN', 'DGRDG', 'FACADV', 'FACBEN', 'FACCHAL', 'FACIND', 'FACLOC', 'FACRESP', 'FACSAL', 'FACSEC']
data_with_dummies = pd.get_dummies(data, columns = column)
data_with_dummies.head()
```

```python
data_with_dummies.to_csv('Data.csv', index=False) #Saving our Clean Dataset in csv format
```

```python
data_with_dummies.dtypes
```

```python
corr_matrix = data_with_dummies.corr().abs()  # Use abs() to consider absolute values for correlation

# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Find index of feature columns with correlation greater than 0.8
to_drop = [column for column in upper.columns if any(upper[column] > 0.8)]

# Drop highly correlated columns
data_reduced = data_with_dummies.drop(columns=to_drop)

# Print remaining columns
print(data_reduced.columns)
```
